```python
import pandas as pd

df = pd.read_csv('store-sales-time-series-forecasting/train.csv').reset_index(drop=True)
```

```python
df=df.drop(columns=['id','onpromotion'])
df
```

```python
df['nbr_family'] = df.apply(lambda x: f"{x['store_nbr']}_{x['family']}", axis=1)
df = df.drop(columns=['store_nbr','family'])
df
```

```python
df_pv = df.pivot(index='date',columns='nbr_family',values='sales').fillna(0)
```

```python
df_pv
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
tr_len = int(0.8*len(df_pv))
tr_data = df_pv.iloc[:tr_len]
te_data = df_pv.iloc[tr_len:]

tr_scale = scaler.fit_transform(tr_data)
te_scale = scaler.fit_transform(te_data)
```

```python
tr_scale
```

```python
import numpy as np

def process_data(data,input_len,out_len):
    x,y = [],[]
    for i in range(len(data)-input_len-out_len+1):
        x.append(data[i:i+input_len][:])
        y.append(data[i+input_len:i+input_len+out_len][:])
    return np.array(x),np.array(y)
```

```python
tr_x,tr_y = process_data(tr_scale,120,16)
te_x,te_y = process_data(te_scale,120,16)
```

```python
tr_x[0:32,-1,:][0]
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
#轉成 tensorflow 可以使用的張量
X_train_tensor = tf.convert_to_tensor(tr_x,dtype = tf.float32)
y_train_tensor = tf.convert_to_tensor(tr_y,dtype = tf.float32)
X_test_tensor = tf.convert_to_tensor(te_x,dtype = tf.float32)
y_test_tensor = tf.convert_to_tensor(te_y,dtype = tf.float32)
# 建立訓練集 Dataset
batch_size = 32
train_dataset = tf.data.Dataset.from_tensor_slices((tr_x, tr_y)) \
                               .shuffle(buffer_size=1000) \
                               .batch(batch_size)

# 建立測試集 Dataset
test_dataset = tf.data.Dataset.from_tensor_slices((te_x, te_y)) \
                              .batch(batch_size)
```

```python
# 使用時序卷積神經網路 (Temporal CN)
def build_tcn_model():
    model = tf.keras.Sequential([
        # 將資料型態由 (batch_size, timesteps, feature) 轉成 (batch_size, feature), timesteps)
        tf.keras.layers.Permute((2, 1)),
        # 接受 (batch_size, feature), timesteps) 之型態，
        tf.keras.layers.Conv1D(64, kernel_size=3, padding='same', dilation_rate=1, activation='relu'),
        tf.keras.layers.Conv1D(64, kernel_size=3, padding='same', dilation_rate=2, activation='relu'),
        tf.keras.layers.Conv1D(64, kernel_size=3, padding='same', dilation_rate=4, activation='relu'),
        tf.keras.layers.Lambda(lambda x: x[:, :, -1]),  # Select the last time step
        tf.keras.layers.Dense(33 * 54 * 16),
        tf.keras.layers.Reshape((16, 33 * 54)),
    ])
    return model

# 加載模型
model = build_tcn_model()

# 編譯模型
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),  # Adam 優化器
              loss='mse',  # 損失函數為 MSE
              metrics=['mse'])  # 評估指標為 MSE

# 訓練模型
epochs = 10
model.fit(train_dataset, validation_data=test_dataset, epochs=epochs)
model.summary()
```

```python
pred_scaler = scaler.fit_transform(df_pv)
pre_data = pred_scaler[-120:]
```

```python
input_tensor = tf.convert_to_tensor(pre_data[tf.newaxis, ...], dtype=tf.float32)
print("模型輸入形狀：", input_tensor.shape)
predictions = model.predict(input_tensor)

# 查看預測結果形狀
print("預測結果形狀：", predictions.shape)
predictions = scaler.inverse_transform(predictions[0])
predictions = np.maximum(predictions, 0)
```

```python

test_df = pd.read_csv('store-sales-time-series-forecasting/test.csv')
test_df['store_family'] = test_df['store_nbr'].astype(str) + '_' + test_df['family']
dates = sorted(test_df['date'].unique())
pred_df = pd.DataFrame(predictions, index=dates, columns=df_pv.columns)
pred_long = pred_df.stack().reset_index()  # opposite of pivot basically
pred_long.columns = ['date', 'store_family', 'sales']

submission = test_df.merge(pred_long, on=['date', 'store_family'], how='left')[['id', 'sales']]
submission.to_csv('submission.csv', index=False)

print("Predictions saved to submission.csv")
```
